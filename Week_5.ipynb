### `Recap` Activity
* Develop an `algorithm` to read `five scores` and `calculate the average`. Then display a
comment, “`Your WAM is over 80`” if the `average is >= 80`, “`Please work harder`” if the `average is < 80`.


### <mark>IPO Chart</mark>
`INPUT`

    *list your input variables here*
    * mark1, mark2, mark3, mark4, mark5

`PROCESS`

    *outline your input process here*
    GET mark1, mark2, mark3, mark4, mark5

    *outline your data-manipulation process here*
    CALCULATE average_score based on the 5 scores

    *outline your output process here*
    DISPLAY "Your WAM is 80 or above." IF average_score is 80 and above
    ELSE DISPLAY "Please work harder"
`OUTPUT`

    *list your output variables here*
    * average_score



### <mark>Pseudocode</mark>
`START`
    
    # Input:
    # GET mark1, mark2, mark3, mark4, mark5
    PROMPT and READ mark1, mark2, mark3, mark4, mark5

    # Process:
    # CALCULATE average_score based on the 5 scores
    average_score = (mark1+mark2+mark3+mark4+mark5)/5

    # Output:
    # DISPLAY "Your WAM is 80 or above." IF average_score is 80 and above
    # ELSE DISPLAY "Please work harder"
    IF average_score >= 80 THEN
        DISPLAY "Your WAM is 80 or over."
    ELSE
        DISPLAY "Please work harder."

`STOP`

```python
def is_valid(mark):
    '''
    Validates and return true if a mark is within the domain of 0 to 100.
    otherwise, return false.
    '''
    if 0 <= mark <= 100:
        return True # mark is valid

    print("Your score must be a number between 0 and 100 only")
    return False # mark is not valid

# Input:
# GET mark1, mark2, mark3, mark4, mark5
# PROMPT and READ mark1, mark2, mark3, mark4, mark5
try:
    mark1, mark2, mark3, mark4, mark5 = \
        map(float, input("Enter five scores separated by a space: ").split())
except:
    print("Scores must be numbers only.")
    mark1, mark2, mark3, mark4, mark5 = 0, 0, 0, 0, 0 # default scores

# if mark1 < 0 or mark2 < 0 or mark3 < 0 or mark4 < 0 or mark5 < 0 or \
#    mark1 > 100 or mark2 > 100 or mark3 > 100 or mark4 > 100 or mark5 > 100:
#     print("Scores must be between 0 and 100.")

if not is_valid(mark1):
    mark1 = 0 # default score
if not is_valid(mark2):
    mark2 = 0 # default score
if not is_valid(mark3):
    mark3 = 0 # default score
if not is_valid(mark4):
    mark4 = 0 # default score
if not is_valid(mark5):
    mark5 = 0 # default score

# print(mark1, mark2, mark3, mark4, mark5) # debug statement

# score count is 5
score_count = 5

# Process:
# CALCULATE average_score based on the 5 scores
# average_score = (mark1+mark2+mark3+mark4+mark5)/5
average_score = sum([mark1, mark2, mark3, mark4, mark5])/score_count

# Output:
# DISPLAY "Your WAM is 80 or above." IF average_score is 80 and above
# ELSE DISPLAY "Please work harder"
if average_score >= 80:
    print(f"Your WAM:{average_score} is 80 or over.")
else:
    print(f"Your WAM:{average_score}. Please work harder.")
```

```python
# @title Activity 4 | sample solution {display-mode: "form"}
# Develop and code an algorithm to read two numbers as inputs and then
# compare and determine if the first number is greater than,
# less than or equal to the second number.

# INPUT: two numbers: num1, num2
# GET num1, num2
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# PROCESS:
# Determine if num1 >, < or = to num2
if num1 > num2: # greater than
    comparison_outcome = f'{num1} is greater than {num2}'

elif num1 < num2: # less than
    comparison_outcome = f'{num1} is less than {num2}'

else: # equal
    comparison_outcome = f'{num1} is equal to {num2}'

# OUTPUT: comparison_outcome
# DISPLAY comparison_outcome
print(comparison_outcome)
```
